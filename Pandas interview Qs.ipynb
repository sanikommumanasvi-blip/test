#Vectorized operations in pandas apply a function to an entire array or column at once, rather than iterating through individual elements using Python loops. This approach leverages highly optimized, pre-compiled C and Fortran code (via the underlying NumPy library), resulting in significantly faster and cleaner code, especially for large datasets. 
Vectorized operations are dramatically faster than explicit Python loops (including for loops, iterrows(), and often apply())

```python

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
# Vectorized addition
df['C'] = df['A'] + df['B']
# Other operations like subtraction, multiplication, division also work
df['D'] = df['A'] * df['B']
df
```

```python
Calculate the mean of a column
mean_value = df['A'].mean()
# Other functions like median, sum, std, min, max are also vectorized
max_value = df['B'].max()
```

```python
import numpy as np
# Create a new column 'Category' based on a condition
df['Category'] = np.where(df['A'] > 2, 'High', 'Low')
df
```

```python
# Boolean indexing
filtered_df = df[df['A'] > 1]
filtered_df
```

```python
# Create a sample DataFrame with missing values
df = pd.DataFrame({'A': [1, 2, np.nan, 4, 5],
                   'B': [6, np.nan, np.nan, 9, 10],
                   'C': [np.nan, 1, 2, np.nan, 3]})

print("Original DataFrame:")
print(df)

# Fill missing values using default linear interpolation
df_linear = df.interpolate()
print("\nDataFrame after linear interpolation:")
print(df_linear)
```
